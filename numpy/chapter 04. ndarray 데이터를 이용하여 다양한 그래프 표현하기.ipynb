```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 그래프 데이터 생성

```python
x=np.linspace(0,10,11)
y=x**2 + x + x+np.random.randn(11)

print(x)
print(y)
```

### 그래프 출력하기
- plot함수(선 그래프), scatter(점 그래프), hist(히스토그램) 등 사용
    - 함수의 parameter 혹은 plt의 다른 함수로 그래프 형태 및 설정을 변경 가능
    - 기본적으로 x,y에 해당하는 값이 필요

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

### 그래프에 주석 추가

- ####  x, y축 및 타이틀

```python
plt.plot(x,y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('x-y relation')
```

- #### grid추가

```python
plt.plot(x,y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('x-y relation')
plt.grid(True)
```

- #### x,y축 범위 지정

```python
plt.plot(x,y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('x-y relation')
plt.grid(True)
plt.xlim(0,20)
plt.ylim(0,200)
```

## plot 함수 parameters

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'#ff00ff')
```

- 그래프 선스타일 변경

```python
plt.plot(x,y,'-.')
```

```python
plt.plot(x,y,'g^')
```

- 그래프 두께 변경
    - linewidth 명시

```python
plt.plot(x,y,linewidth=7)
```

- keyword parameter 이용하여 모든 속성 설정
    - https://matplotlib.org/api/pyplot_api.html#matplotlib.pyploy.plot
    - color
    - linestyle
    - marker
    - markerfacecolor
    - markersize 등등

```python
plt.plot(x,y,color='black',
        linestyle='--',marker='^',
        markerfacecolor='blue', markersize=10)
```

## subplot으로 여러 그래프 출력하기
- subplot함수로 구획을 구별하여 각각의 subplot에 그래프 출력

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(x,y,'g')

plt.subplot(2,2,3)
plt.plot(y,x,'k')

plt.subplot(2,2,4)
plt.plot(x,np.exp(x),'b')
```

## hist함수 사용
- histogram 생성
- bins로 histogram bar 개수 설정

```python
data=np.random.randint(1,100,size=200)
print(data)
```

```python
plt.hist(data,bins=10)
plt.xlabel('값')
plt.ylabel('개수')
plt.grid(True)
```

```python
plt.hist(data,bins=20,alpha=0.3)
plt.xlabel('값')
plt.ylabel('개수')
plt.grid(True)
```

2. numpy를 이용하여 pi(원주율) 값을 계산하시오
  - 몬테 카를로 방법 이용하기
  - ![파이 값 구하기](http://mathfaculty.fullerton.edu/mathews/n2003/montecarlopi/MonteCarloPiMod/Images/MonteCarloPiMod_gr_25.gif)
  - 이미지 출처: http://mathfaculty.fullerton.edu/mathews/n2003/montecarlopi/MonteCarloPiMod/Images/MonteCarloPiMod_gr_25.gif 

pi/4:1=(4분원안에 생성된 점 개수) : 전체 시도 횟수

pi=4*(4분원 안에 생성된 점 개수) / 전체시도횟수(1e7)

```python
total=int(1e7)

points=np.random.rand(total,2)
4*np.sum(np.sum(points**2,axis=1)<1)/total
```
