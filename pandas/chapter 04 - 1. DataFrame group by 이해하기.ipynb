```python
import pandas as pd
import numpy as np
```

#### group by
- 아래의 세 단계를 적용하여 데이터를 그룹화(grouping) (SQL의 group by와 개념적으로는 동일, 사용법은 유사)
    - 데이터 분할
    - operation 적용
    - 데이터 병합

```python
df=pd.read_csv('C://Users/whs38//OneDrive//바탕 화면//archive//titanic.csv')
df.head()
```

#### Groupby groups 속성
- 각 그룹과 그룹에 속한 index를 dict형태로 표현

```python
class_group=df.groupby('Pclass')
class_group
```

```python
class_group.groups#각 값에 해당하는 인덱스가 나옴
```

```python
gender_group=df.groupby('Sex')
gender_group
```

#### grouping 함수
- 그룹 데이터에 적용 가능한 통계 함수 (NaN은 제외하여 연산)
- count - 데이터 개수
- sum - 데이터의 합
- mean.std,var - 평균, 표준편차, 분산
- min, max - 최소, 최대값

```python
class_group.count()
```

```python
class_group.sum()
```

```python
class_group.mean()['Age']
```

- 성별에 따른 생존률 구해보기

```python
df.groupby('Sex')['Survived'].mean()
```

#### 복수 columns로 grouping 하기
- groupby에 column 리스트를 전달
- 통계함수를 적용한 결과는 multiindex를 갖는 dataframe

- 클래스와 성별에 따른 생존률 구해보기

```python
df.groupby(['Pclass','Sex'])['Survived'].mean()
```

#### index를 이용한 group by
- index가 있는 경우, groupby 함수에 level 사용 가능
    - level은 index의 depth를 의미하며, 가장 왼쪽부터 0부터 증가

- **set_index 함수**
    - column 데이터를 index 레벨로 변경
- **reset_index 함수**
    - 인덱스 초기화

```python
df.head()
```

```python
df.set_index('Pclass')
```

```python
df.set_index(['Pclass','Sex'])
```

```python
df.set_index(['Pclass','Sex']).reset_index()
```

```python
df.set_index('Age').groupby(level=0).mean()
```

#### Age컬럼을 인덱스로 하고 나이대별 생존률 구하기

```python
import math
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age/10)*10
```

```python
df.set_index('Age').groupby(age_categorize).mean()#인덱스가 있는 경우에만 사용가능(함수가 자동적으로 인덱스의 값들을 변수로 받음????)
```

#### Multiindex를 이용한 grouping

```python
df.set_index(['Pclass','Sex']).groupby(level=[0,1]).mean()
```

#### aggregate(집계) 함수 사용하기
- groupby 결과에 집계함수를 적용하여 그룹별 데이터 확인가능

```python
df.set_index(['Pclass','Sex']).groupby(level=[0,1]).aggregate(['mean','sum','max'])
```
