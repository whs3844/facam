#### row 선택하기
- Series의 경우 []로 row 선택이 가능하나, DataFrame의 경우는 기본적으로 column을 선택하도록 설계
- .iloc, .loc로 row 선택가능
    - loc - 인덱스 자체를 사용
    - iloc - 0 based index로 사용
    - 이 두 함수는 ,를 사용하여 column 선택도 가능

```python
train_data.index=np.arange(100,991)
```

```python
train_data.loc[986]
```

```python
train_data.loc[[100,110,990]]
```

```python
train_data.iloc[0]
```

#### row, column 동시에 선택하기
- loc,iloc 속성을 이용할 때 

```python
train_data.loc[[100,110,990],['PassengerId','Survived','Sex']]
```

```python
train_data.iloc[[101,100,200,102],[0,1,2]]
```
